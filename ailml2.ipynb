```python
from collections import deque
```

```python
graph={
    'A':["B","C"], #child of A is B and C
    "B":["D","E"],
    "C":["F"],
    "D":[],
    "E":[],
    "F":[]
}
```

```python
print(graph)
```

```python
q= deque()
q.append('A')
print(q)
q.append('B')
print(q)
print(q.popleft())

```

```python
start="A"
goal="F"
```

```python
queue= deque([start])
visited= set([start])
parent= ()
```

```python
print(queue)
```

```python
parent={start:None}
print(parent)
```

```python
step = 0
while queue:
  print("currnet queue", list(queue))
  input("press enter for next step")
  step += 1
  node= queue.popleft()

  print("/n....")
  print("step", step)
  print("popped node", node)
  print("queue after pop", list(queue))

  if node==goal:
    print("goal found")
    break
```

```python
for neighbour in graph[node]:
  print(f" checking neighbor {neighbor}of {node}...")
  print("now processing neighbor : ",neighbor)
  if neighbor not in visited:

    visited.add(neighbor)
    parent[neighbor]=node
    queue.append(neighbor)

    print(" parent[( )]={}",format(neighbor,node))
    print("added to queue: ",neighbor)

  else:
    print("already")
    #print("current queue: ",list[queue])

```
